```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

infos = []
details = []
for page in range(1, 8000):
    if page == 2839 :
        continue
    if page == 3079 :
        continue
    url = 'https://movies.yahoo.com.tw/movieinfo_main/%s' % page
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    #----------#
    for d in soup.find_all('div', class_="movie_intro_info _c"):
        info = {}
        info[''] = d.find('div', class_="movie_intro_info_r").find('h1').text.replace(' ', '').replace('\n', '')
        info[''] = d.find('div', class_="movie_intro_info_r").find('a', class_='gabtn').text.replace(' ', '').replace('\n', '')
        info[''] = d.find('div', class_='movie_intro_info_r').find('span').text.replace('', '').replace(' ', '')
        infos.append(info)
    #----------#
    for d in soup.find_all('div', class_='gray_infobox storeinfo'):
        detail = {}
        detail[''] = d.find('div', class_="gray_infobox_inner").find('span').text.replace(' ', '').replace('\n', '').replace('\r\xa0', '').replace('\r', '')
        details.append(detail)
    #----------#
results = []
for info, detail in zip(infos, details):
    results.append(dict(list(info.items()) + list(detail.items())))
print(results)

#-----movie.csv-----#
df = pd.DataFrame(results)
df.to_csv('movies.csv', encoding='utf_8_sig')
```

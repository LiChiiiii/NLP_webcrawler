```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

check = 0
infos = []
genres = []
for genreID in range (1,20):
    genre_url = 'https://movies.yahoo.com.tw/moviegenre_result.html?genre_id=%s' % genreID
    for page in range(1,442):
        url = genre_url+'&page=%s' % page
        print(url)
        r = requests.get(url)
        soup = BeautifulSoup(r.text)
        #-----相關資訊-----#
        for d in soup.find_all('div', class_="release_info"):
            info = {}
            info['名稱'] = d.find('div', class_="release_movie_name").find('a', class_='gabtn').text.replace(' ', '').replace('\n', '')
            info['分類'] = soup.title.text.replace('類電影 - Yahoo奇摩電影', '')
            info['上映日期'] = d.find('div', class_='release_movie_time').text.replace('上映日期 ： ', '')
            info['劇情介紹'] = d.find('div', class_='release_text').find('span').text.replace(' ', '').replace('\n', '').replace('\r\xa0', '').replace('\r', '')
            
            for s in infos :
                if info['名稱'] == s.get('名稱'):
                    check = 1
                    break
                else : check = 0
            
            if check == 0:
                infos.append(info)

#-----存到movie.csv-----#
df = pd.DataFrame(infos)
df.to_csv('movies.csv', encoding='utf_8_sig')
```
